```python
1.Write a Python program to find sum of elements in list?
Ans:-
def sumlist(list):
    sum=0
    for i in range(len(list)):
        sum = sum+list[i]
    return sum

#initialise list
list = [10, 9, 7, 5]
print(list)
print("sum of list: ",sumlist(list))

Output:-
    [10, 9, 7, 5]
    
2. Write a Python program to Multiply all numbers in the list?
Ans:-
def multiply_list(items):
    tot = 1
    for x in items:
        tot *= x
    return tot
print(multiply_list([1,2,-8]))

Output:-
    -16
    
3.Write a Python program to find smallest number in a list?
Ans:-
# Python program to find smallest
# number in a list
 
# list of numbers
list1 = [10, 20, 4, 45, 99]
 
# sorting the list
list1.sort()
 
# printing the first element
print("Smallest element is:", list1[0])

Output:-
    Smallest element is: 4

4.Write a Python program to find largest number in a list?
Ans:-
# Python program to find largest
# number in a list
 
# list of numbers
list1 = [10, 20, 4, 45, 99]
 
# sorting the list
list1.sort()
 
# printing the last element
print("Largest element is:", list1[-1])

Output:-
    Largest element is: 99
        
5. Write a Python program to find second largest number in a list?
Ans:-
# Python program to find largest number
# in a list
 
# List of numbers
list1 = [10, 20, 20, 4, 45, 45, 45, 99, 99]
 
# Removing duplicates from the list
list2 = list(set(list1))
 
# Sorting the  list
list2.sort()
 
# Printing the second last element
print("Second largest element is:", list2[-2])

Output:-
    Second largest element is: 45

6. Write a Python program to find N largest elements from a list?
Ans:-
    # Python program to find N largest
# element from given list of integers
 
# Function returns N largest elements
def Nmaxelements(list1, N):
    final_list = []
 
    for i in range(0, N):
        max1 = 0
         
        for j in range(len(list1)):    
            if list1[j] > max1:
                max1 = list1[j];
                 
        list1.remove(max1);
        final_list.append(max1)
         
    print(final_list)
 
 # Driver code
list1 = [2, 6, 41, 85, 0, 3, 7, 6, 10]
N = 2
 
# Calling the function
Nmaxelements(list1, N)

Output:-
    [85, 41]
    

7. Write a Python program to print even numbers in a list?
Ans:-
    #even numbers in list

#function
def even(list):
    new_list=[]
    for i in list:
        if i%2==0:               
            new_list.append(i)
    return new_list

#input
li=[]
n=int(input("Enter size of list "))
for i in range(0,n):
    e=int(input("Enter element of list "))
    li.append(e)

print("Even numbers in ",li)
print(even(li))

Output:-
Enter size of list  5
Enter element of list  6
Enter element of list  7
Enter element of list  28
Enter element of list  45
Enter element of list  515
Even numbers in  [6, 7, 28, 45, 515]
[6, 28]

8. Write a Python program to print odd numbers in a List?
Ans:-
# Python program to print odd Numbers in a List
 
# list of numbers
list1 = [10, 21, 4, 45, 66, 93]
 
# iterating each number in list
for num in list1:
 
    # checking condition
    if num % 2 != 0:
    print(num, end=" ")
    
Output:-
    21 45 93 
    
9. Write a Python program to Remove empty List from List?
Ans:-
# Python3 code to Demonstrate Remove empty List
 
# Initializing list
test_list = [5, 6, [], 3, [], [], 9]
 
# printing original list
print("The original list is : " + str(test_list))
 
# Remove empty List from List
while [] in test_list :
    test_list.remove([])
 
 #printing result
print("List after empty list removal : " + str(test_list))

Output:-
    
The original list is : [5, 6, [], 3, [], [], 9]
List after empty list removal : [5, 6, 3, 9]
    
    
10. Write a Python program to Cloning or Copying a list?
Ans:-
li=[]
n=int(input("Enter size of list "))
for i in range(0,n):
    e=int(input("Enter element of list "))
    li.append(e)
print("Original list: ",li)

#cloning
list_copy = li[:]
print("After cloning: ",list_copy)

Output:-
Enter size of list 5
Enter element of list 3
Enter element of list 6
Enter element of list 1
Enter element of list 2
Enter element of list 3
Original list: [3, 6, 1, 2, 3]
After cloning: [3, 6, 1, 2, 3]
    
11. Write a Python program to Count occurrences of an element in a list?
Ans:-
sample_list = ["a", "ab", "a", "abc", "ab", "ab"]
print(sample_list.count("a"))
print(sample_list.count("ab")
      
Output:-
2
3
 
```
